```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Total_fertility_rate"
page = requests.get(url)
soup = BeautifulSoup(page.text , "html")
print(soup)
```

```python
soup.find("table")
```

```python
table = soup.find("table" , class_ = "wikitable mw-collapsible")
```

```python
world_titles = table.find_all("td")
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
```

```python
x = world_table_titles[1:3]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = x)
df
```

```python
y = world_table_titles[3 :: 2 ]
y
```

```python
z = world_table_titles[4::2]
z
```

```python
d = {
    'Years': y,
    'TFR': z,
    
}
df = pd.DataFrame(data=d)
df
```

```python
import matplotlib.pyplot as plt 
dd = pd.DataFrame(d)
  
plt.plot(dd['TFR'], dd["Years"], color='red', marker='o')
plt.title('Fertility rate', fontsize=14)
plt.xlabel('TFR', fontsize=14)
plt.ylabel('YEARS', fontsize=14)
plt.xlim(max(z), min(z))
plt.ylim(max(y), min(y))

plt.show()
```
